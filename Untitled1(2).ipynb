```python

from qiskit import QuantumCircuit, transpile
from qiskit.visualization import *
from qiskit import *
from qiskit_aer import Aer
import numpy as np
from qiskit.visualization import plot_histogram
from qiskit.circuit.library import UnitaryGate
```

```python
def errores_Amplitud(n, q):
    """
    Crea una matriz 2^n x 2^n que aplica una puerta X al qubit q.

    Parámetros:
    n: int
        Número total de qubits en el sistema.
    q: int
        Índice del qubit al que se aplicará la puerta X (0-indexado).

    Retorna:
    numpy.ndarray
        Matriz 2^n x 2^n que representa la transformación.
    """
    X = np.array([[0, 1], [1, 0]])  # Puerta X para un solo qubit
    I = np.eye(2)  # Identidad 2x2

    # Construcción de la matriz usando producto de Kronecker
    matrices = [I] * n
    matrices[q] = X  # Aplicar X en el qubit especificado

    gate_matrix = matrices[0]
    for mat in matrices[1:]:
        gate_matrix = np.kron(gate_matrix, mat)

    return gate_matrix
```

```python
def correccion_errores_Amplitud(n,result):

    if result == "00":
        return np.eye(2**n)  # No hacer nada (devolver identidad)
    elif result == "10":
        return errores_Amplitud(n, 1)  # Aplicar X en el qubit 1
    elif result == "01":
        return errores_Amplitud(n, 2)  # Aplicar X en el qubit 2
    elif result == "11":
        return errores_Amplitud(n, 0)  # Aplicar X en el qubit 0

    return np.eye(2**n)
```

```python
n=5
qc = QuantumCircuit(n,3)

#qc.x(0)
qc.barrier()

#Aplicar las puertas H a los qubits
for i in range (n-3):
  qc.cx(0,i+1)

qc.barrier()

#Error
error_matrix = errores_Amplitud(n-2, 0) # Genera un error de amplitud en un solo qubit (qubit del segundo parámetro)
error_gate = UnitaryGate(error_matrix, label = "Error")
qc.append(error_gate, range(n-2))
qc.barrier()

# Calcular el síndrome
qc.cx(0,3)
qc.cx(1,3)

qc.cx(0,4)
qc.cx(2,4)

qc.barrier()

# Mediciones
qc.measure(3,0)
qc.measure(4,1)

qc.barrier()

#resultados
sim = Aer.get_backend('qasm_simulator')

my_circuit = transpile(qc, sim)
job1 = sim.run(my_circuit)

result = job1.result()
counts = result.get_counts()
print("Resultados:", counts)

# Correción de errores
correction_matrix = correccion_errores_Amplitud(n-2,list(counts.keys())[0])
#print(list(counts.keys())[0])
error_correct_gate = UnitaryGate(correction_matrix, label = "Error_Correct")
qc.append(error_correct_gate, range(n-2))
qc.barrier()

# Medimos los qubits

for i in range(n-2):
  qc.measure(i,i)

sim1 = Aer.get_backend('qasm_simulator')

my_circuit1 = transpile(qc, sim1)
job2 = sim.run(my_circuit)
result1 = job1.result()
counts1 = result1.get_counts()
print("Resultados despues de correcion:", counts1)
```
