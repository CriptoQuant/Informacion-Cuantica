```python
from qiskit import QuantumCircuit, transpile
from qiskit.visualization import *
from qiskit import *
from qiskit_aer import Aer
import numpy as np
from qiskit.visualization import plot_histogram
from qiskit.circuit.library import UnitaryGate, XGate
import random
from qiskit_aer import StatevectorSimulator
```

```python
n = 7  # Número de qubits físicos
qc = QuantumCircuit(2 * n - 1, n-1)  # 7 qubits de código + 6 ancilla

#qc.x(7)
#qc.z(7)

# Inicialización: aplicar Hadamard en las ancillas
for i in range(6):
    qc.h(i)
qc.barrier()


# Detección errores X

qc.cx(0,[i+5 for i in [4,5,6,7]]) # Controlled g1
qc.barrier()
qc.cx(1,[i+5 for i in [2,3,6,7]]) # Controlled g2
qc.barrier()
qc.cx(2,[i+5 for i in [1,3,5,7]]) # Controlled g3
qc.barrier()

# Detección errores Z
qc.cz(3,[i+5 for i in [4,5,6,7]]) # Controlled g4
qc.barrier()
qc.cz(4,[i+5 for i in [2,3,6,7]]) # Controlled g5
qc.barrier()
qc.cz(5,[i+5 for i in [1,3,5,7]]) # Controlled g6
qc.barrier()

for i in range(6):
    qc.h(i)
qc.barrier()

for i in range(6):
    qc.measure(i,i)

#qc.z(7)
```

```python
sim = Aer.get_backend('qasm_simulator')

my_circuit = transpile(qc, sim)
job1 = sim.run(my_circuit)

result = job1.result()
counts = result.get_counts()
print("Resultados:", counts)


simulator = StatevectorSimulator()
job2 = simulator.run(my_circuit)  # Run the circuit using the statevector simulator
statevector = job2.result().get_statevector()
print("Estado del sistema:", statevector)

# Now you can access the amplitudes of the statevector
print('Number of kets in superposition: ', len(statevector))
print("Kets in superposition:")
for i, amp in enumerate(statevector):
    if abs(amp) > 1e-6:  # Filter out negligible amplitudes
        print('Ket:', bin(i)[2:].zfill(qc.num_qubits), '--- Amplitude:', amp)
```

```python
# Dibujar el circuito
qc.draw('mpl')
```
